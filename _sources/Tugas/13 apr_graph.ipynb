```python
import networkx as nx 
import matplotlib.pyplot as plt
```

```python
G=nx.Graph()
```

```python
G.add_nodes_from([(1,2,3,4,5,6,7,8,9)])
```

```python
G.add_edges_from([(1,2),(1,3),(1,4),(2,3),(3,4),(4,5),(4,6),(5,6),(5,7),(5,8),(6,8),(7,8),(7,9)])
nx.draw(G,with_labels=True)
```
