````python
import twint
import nest_asyncio # assynchronous 
nest_asyncio.apply()

c = twint.Config()
c.Limit = 500
c.Search = "Ridwan Kamil" # topik
c.Pandas = True
c.Store_csv = True
twint.run.Search(c)
Tweets_df = twint.storage.panda.Tweets_df # twint menyimpan data dalam bentuk data frame dibantu dengan pandas
````
